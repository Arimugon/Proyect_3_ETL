# B. Web Scraping con Selenium de Wikipedia:

El siguiente método que utilizo es web scraping con selenium de wikipedia para coger una tabla de wikipedia en la que se ven reflejados los ganadores de división de honor femenina y masculina de los últimos 10 años. Lo que quiero ver con esta tabla es que la mayoría de los clubs con escula son los que llevan a división de honor.

```python
%pip install webdriver-manager
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

url = 'https://es.wikipedia.org/wiki/Divisi%C3%B3n_de_Honor_femenina_de_rugby'
PATH=ChromeDriverManager().install()   

driver=webdriver.Chrome(PATH)         


```

```python
url = 'https://es.wikipedia.org/wiki/Divisi%C3%B3n_de_Honor_femenina_de_rugby'
driver=webdriver.Chrome(PATH)

driver.get(url)
```

```python
tabla=driver.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[4]/tbody')
tabla
```

```python
data=[[e.text for e in f.find_elements(By.TAG_NAME, 'td')] 
      for f in tabla.find_elements(By.TAG_NAME, 'tr')]
data[:2]
```

```python
cabeceras=driver.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[4]/tbody/tr[1]')

cabeceras=[c.text for c in cabeceras.find_elements(By.TAG_NAME, 'th')]



cabeceras
```

```python
df=pd.DataFrame(data, columns=cabeceras)

df
```

```python
df.drop([0], axis=0, inplace=True)
df
```

```python
df.at[4,'Año']='2013-14'
```

```python
df = df.reset_index(drop=True)
```
